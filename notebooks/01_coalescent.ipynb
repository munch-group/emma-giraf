# Coalescent og state space

I denne notebook konstruerer jeg state space for coalescent-processen ved hjælp af Phasic.

Coalescent-modellen beskriver, hvordan genetiske linjer smelter sammen bagud i tid, indtil vi når den fælles stamfader (MRCA).
Hver state repræsenteres som en vektor, der tæller hvor mange linjer der har et bestemt antal efterkommere.

```python
from phasic import Graph, with_ipv
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
nr_samples = 4

@with_ipv([nr_samples]+[0]*(nr_samples-1))
def coalescent(state):
    transitions = []

    for i in range(state.size):
        for j in range(i, state.size):
            same = int(i == j)

            if same and state[i] < 2:
                continue
            if not same and (state[i] < 1 or state[j] < 1):
                continue

            new = state.copy()
            new[i] -= 1
            new[j] -= 1
            new[i+j+1] += 1

            rate = state[i]*(state[j]-same)/(1+same)
            transitions.append((new, rate))

    return transitions
```

```python
graph = Graph(coalescent)
graph.plot()
```

```python
states = graph.states()
print(states)
```

Hver state repræsenterer:

- indeks 0: antal singletons
- indeks 1: antal doubletons
- osv.

Processen starter med alle linjer som singletons og slutter med en lineage (MRCA).
