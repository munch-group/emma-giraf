# Rewards og Site Frequency Spectrum (SFS)

Jeg bruger rewards til at beregne biologisk relevante størrelser såsom branch lengths og SFS.

```python
graph = Graph(coalescent)

reward_matrix = graph.states().T
labels = [f"{i+1}'ton" for i in range(reward_matrix.shape[0])]

sfs = [graph.expectation(r) for r in reward_matrix]

plt.figure(figsize=(5,3))
sns.barplot(x=labels, y=sfs)
plt.title("Site Frequency Spectrum")
plt.show()
```

SFS beskriver hvor meget genetisk variation der findes i forskellige frekvenser i data.

Dette er centralt i population genetics og bruges til at sammenligne modeller med data.
