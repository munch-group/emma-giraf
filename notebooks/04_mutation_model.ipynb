# Mutationer i coalescent-modellen

Her udvider jeg modellen ved at inkludere mutationer som diskrete events langs grenene.

```python
mutation_graph = Graph(coalescent)

def mutation_rate(state, rate):
    return sum(state) * rate

mutation_rate_value = 0.1

rewards = mutation_graph.discretize(mutation_rate, mutation_rate=mutation_rate_value)
```

## Forventet antal mutationer

```python
mutation_graph.expectation(rewards)
```

```python
rt_graph = mutation_graph.reward_transform(rewards)

x = np.arange(10)
plt.bar(x, rt_graph.pdf(x))
plt.title("Distribution of mutations")
plt.show()
```

Mutationer modelleres som en proces langs grenene, hvor antallet afhænger af den samlede grenlængde.

Dette forbinder modellen direkte til genetiske data.
